```python
from random import randint


table = [
    {"name" : "Арсенал", "games" : 0, "points" : 0, "wins" : 0, "loses" : 0, "draws" : 0, "goals" : 0, "misses" : 0},
    {"name" : "Челси", "games" : 0, "points" : 0, "wins" : 0, "loses" : 0, "draws" : 0, "goals" : 0, "misses" : 0},
    {"name" : "Лестер", "games" : 0, "points" : 0, "wins" : 0, "loses" : 0, "draws" : 0, "goals" : 0, "misses" : 0},
    {"name" : "Эвертон", "games" : 0, "points" : 0, "wins" : 0, "loses" : 0, "draws" : 0, "goals" : 0, "misses" : 0},
    {"name" : "Манчестер сити", "games" : 0, "points" : 0, "wins" : 0, "loses" : 0, "draws" : 0, "goals" : 0, "misses" : 0}
    ]

scores = []

def create_table():
    k = 0
    for i in range (k, len(table)):
        for j in range (k + 1, len(table)):
            goals1 = randint(0, 9)
            goals2 = randint(0, 9)
            if goals1 > goals2:
                table[i]["wins"] += 1
                table[j]["loses"] += 1

                table[i]["points"] += 3

            elif goals1 < goals2:
                table[j]["wins"] += 1
                table[i]["loses"] += 1

                table[j]["points"] += 3

            else:
                table[i]["draws"] += 1
                table[j]["draws"] += 1

                table[i]["points"] += 1
                table[j]["points"] += 1


            table[i]["goals"] += goals1
            table[j]["goals"] += goals2

            table[i]["misses"] += goals2
            table[j]["misses"] += goals1

            table[i]["games"] += 1
            table[j]["games"] += 1

            scores.append([table[i]["name"], table[j]["name"], goals1, goals2])
        k += 1


def sort_table():
    for i in range(len(table)):
        for j in range(len(table)):
            if table[i]["points"] > table[j]["points"]:
                table[i], table[j] = table[j], table[i]

def search_one_team(scores, team):
    flag = 0
    for i in range(len(scores)):
        if team in scores[i]:
            print('{:1s} - {:1s}   {:2d} - {:2d}'.format(scores[i][0], scores[i][1], scores[i][2], scores[i][3]))
            flag = 1

    if flag == 0:
        print("No such team")

def search_two_teams(scores, team1, team2):
    flag = 0
    for i in range(len(scores)):
        if team1 in scores[i] and team2 in scores[i]:
            print('{:1s} - {:1s}   {:2d} -{:2d}'.format(scores[i][0], scores[i][1], scores[i][2], scores[i][3]))
            flag = 1
    if flag == 0:
        print("Team1 or team2 does not exist")

def print_table():
    print(" #   Team           Games      Points      Wins     Loses      Draws      Goals      Misses")
    for i in range(len(table)):
        print('{:2d} {:15s} {:5d} {:10d} {:10d} {:10d} {:10d} {:10d} {:10d}'. format(i + 1,
                                                                         table[i]["name"],
                                                                         table[i]["games"],
                                                                         table[i]["points"],
                                                                         table[i]["wins"],
                                                                         table[i]["loses"],
                                                                         table[i]["draws"],
                                                                         table[i]["goals"],
                                                                         table[i]["misses"]))


create_table()
sort_table()
print_table()

choice = 1
while choice != 0:
    print(
"""
Make your choice:

1 - Search by one team
2 - Search by two teams

0 - Exit
"""
        )
    
    choice = input()
    if choice == 0:
        break
    elif choice == '1':
        team = input("Input team: ")
        search_one_team(scores, team)
    elif choice == '2':
        team1 = input("Input first team: ")
        team2 = input("Input second team: ")
        search_two_teams(scores, team1, team2)
    else:
        print("Wrong input!")
            
        
```
